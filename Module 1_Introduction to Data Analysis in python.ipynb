### 1-1 Read Data:

We use "pandas.read_csv()" function to read the csv file. In the bracket, we put the file path along
with a quotation mark, so that pandas will read the file into a data frame from that address. The
file path can be either an URL or your local file address. Because the data does not include headers,
we can add an argument " headers = None" inside the "read_csv()" method, so that pandas will
not automatically set the first row as a header. You can also assign the dataset to any variable you
create.

```python
import pandas as pd
```

```python
file_name = 'imports-85_csv'
```

```python
df = pd.read_csv(file_name, header=None)
```

After reading data from and store it into  a data frame which we call it df here we can show the top 5 rows using df.head(). Also it is possible to send i parameter as input to df.head(n) to get first n rows.
There is also possible to get buttom rows using df.tail(n).
